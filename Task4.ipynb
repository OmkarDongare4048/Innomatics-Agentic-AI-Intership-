1.Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("âŒ Minimum recharge amount is â‚¹50.")
        return False

    if amount not in valid_plans:
        print("âŒ Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False

    return True


# Main Program
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: â‚¹"))

        if validate_recharge(recharge_amount):
            print("âœ… Recharge Successful!")
            break
        else:
            print("ðŸ” Please try again.\n")

    except ValueError:
        print("âŒ Please enter a valid numeric amount.\n")
```

2.Inventory Reorder Alert System

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product} â†’ Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product} â†’ Stock OK (Stock: {quantity})")


# Example dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18,
    "USB Cable": 8
}

# Function Call
check_inventory(inventory)
```

3.Student Result Processing System

```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    # Determine result
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result


# Example marks list
student_marks = [60, 45, 70, 55, 40]

avg, status = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", status)
```

4.Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare


# Main Program with Retry Option
while True:
    try:
        km = float(input("Enter distance in km: "))

        if km <= 0:
            print("âŒ Distance must be greater than 0.\n")
            continue

        peak_hour = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(km, peak_hour)

        print(f"âœ… Total Fare: â‚¹{total_fare:.2f}")
        break

    except ValueError:
        print("âŒ Please enter valid numeric distance.\n")
```

5. Employee Attendance Eligibility Checker

```python
def check_attendance(attendance_list):
    present_count = 0

    # Count present days using loop
    for day in attendance_list:
        if day.upper() == "P":
            present_count += 1

    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Example attendance record
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

percent, status = check_attendance(attendance)

print(f"Attendance Percentage: {percent:.2f}%")
print("Status:", status)
```

6. Password Strength Checker

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check minimum length
    if len(password) < 8:
        return "âŒ Weak Password (Minimum length is 8)"

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    # Final validation
    if has_digit and has_special:
        return "âœ… Strong Password"
    else:
        return "âŒ Weak Password (Must contain at least one digit and one special character @#$)"


# Example usage
user_password = input("Enter your password: ")
result = check_password_strength(user_password)
print(result)
```
